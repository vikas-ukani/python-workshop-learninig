# HeatMap Graph generate


```python

# Function for create heatmap graph
def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):
    if not ax:
        ax = plt.gca()
    im = ax.imshow(data, **kwargs)
    
    cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom")

    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_yticks(np.arange(data.shape[0]))
    ax.set_xticklabels(col_labels)
    ax.set_yticklabels(row_labels)

    ax.tick_params(top=True, bottom=False, labeltop=True, labelbottom=False)
    
    plt.setp(ax.get_xticklabels(), rotation=0, ha="right", rotation_mode="anchor")

    for edge, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False)
    
    return im, cbar


data = np.array([
    [30, 20, 10,],
    [10, 50, 15], 
    [12, 10, 20]
]  )
im, cbar = heatmap(data, ['Class-1', 'Class-2', 'Class-3'], ['A', 'B', 'C'], cmap='YlGn', cbarlabel='Number of Students')
```



# Generating 3D plots to Plot a Sine Wave



```python
from mpl_toolkits.mplot3d import Axes3D

X = np.linspace(0, 10, 50)
Y = np.linspace(0, 10, 500)
X, Y = np.meshgrid(X, Y)
Z = (np.sin(X))
# Setup axis
fig = plt.figure(figsize=(10,5))


ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
# Add title and axes labels
ax.set_title("Testing of 3D Plot", size=10)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')




plt.show()
```
